The dateset contains rating for every talk, for categorys like Funny, Courageous, Informative, etc. They come in the form of a string encoding a lsit of dictionaries. One dictionary for each category. These string need to be decoded.

```python
ast.literal_eval(data["ratings"][0])
```

This decoding can be done for every talk.

```python
data["ratings"] = data["ratings"].apply(lambda x: ast.literal_eval(x))
```

The lists of dictionarys can be converted to dataframes, and shaped into a easily usable form, this way:

```python
pd.DataFrame(data["ratings"][0]).set_index("name").T
```

This is done for every talk.

```python
data["ratings"] = data["ratings"].apply(lambda x: pd.DataFrame(x).set_index("name").T)
```

Now every talk has a dataframe in its ratings column. This can be accessed easily, to generate ne colums in the data dataframe containing the counts for each category for each talk.

We first need a list of all categories.

```python
rating_categories = data["ratings"][0].columns.tolist()
print(rating_categories)
```

This list can be used to be iterated over to generate the new columns, containg the counts for the categories for each talk.

```python
for category in rating_categories:
    data[category] = data["ratings"].apply(lambda x: x[category]["count"])
```
